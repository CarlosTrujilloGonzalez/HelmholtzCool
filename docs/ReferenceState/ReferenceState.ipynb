```python
from numpy import genfromtxt
from scipy import stats
import matplotlib.pyplot as plt

ref=genfromtxt('Carbondioxide_Short_IIR.csv',delimiter=';',names=True)

# IIR
# s0=1e3
# h0=2e5

# ASHRAE or NBP
s0=0
h0=0
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2, sharey=False)
ax[0].plot(ref['idealPower1'], (ref['sref']-s0), marker='x', color='b')
ax[0].grid(b=True, linestyle=':')
ax[1].plot(ref['idealPower2'], (ref['href']-h0), marker='x', color='r')
ax[1].grid(b=True, linestyle=':')

plt.show()
```

```python
# entropy
slope, intercept, r_value, p_value, std_err = stats.linregress(ref['idealPower1'], ref['sref'])
x_s = (s0-intercept)/slope
print("entropy slope: {}".format(slope))
print("entropy intercept: {}".format(intercept))
print("entropy linregress r_value: {}".format(r_value))
print('{:016.15f}'.format(x_s))

# enthalpy
slope, intercept, r_value, p_value, std_err = stats.linregress(ref['idealPower2'], ref['href'])
x_h = (h0-intercept)/slope
print("\nenthalpy slope: {}".format(slope))
print("enthalpy intercept: {}".format(intercept))
print("enthalpy linregress r_value: {}".format(r_value))
print('{:016.15f}'.format(x_h))
```
